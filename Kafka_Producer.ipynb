```python
import pandas as pd
from kafka import KafkaProducer
import time
from json import dumps
import json

from stockPriseGetter import stockGetPrice
```

```python
producer = KafkaProducer(bootstrap_servers=['54.144.131.204:9092'],
                         value_serializer=lambda x: 
                         dumps(x).encode('utf-8'))
```

```python
while True:
    stockGetPriceObject = stockGetPrice()
    stockPrices = stockGetPriceObject.getPrice()
    
    # Send the entire stockPrices object as a single message
    producer.send('demotest', value=stockPrices)
    
    time.sleep(5)
    print('After Sleep')
```

```python
producer.flush()
```
